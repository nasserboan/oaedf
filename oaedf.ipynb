```python
import pandas as pd
import folium as fl
from folium.plugins import MarkerCluster
```

```python
data = pd.read_csv('data/2018-07-16-oae.csv',sep=';',encoding='latin-1')
```

```python
data
```

```python
data.rename(columns={'COD. OAE':'cod',
                    'DESCRITIVO':'desc',
                    'TIPO DE OAE':'tipo',
                    'LATITUTE':'lat',
                    'LONGITUDE':'long'
                    }, inplace=True)
```

```python
data.info()
```

```python
def replacepontos(df):
    string = df['lat'].replace('.','')
    return float(string[:3]+'.'+string[3:])

data['lat'] = data.apply(replacepontos,axis=1)
```

```python
def replacepontoslong(df):
    string = df['long'].replace('.','') 
    return float(string[:3]+'.'+string[3:])

data['long'] = data.apply(replacepontoslong,axis=1)
```

```python
data.info()
```

```python
m = fl.Map(
    location=[-15.717,-47.865])
```

```python
m
```

```python
marker_cluster = MarkerCluster().add_to(m)

for i in range(0,len(data)):
    popup = 'Tipo: '+data.iloc[i]['tipo']+'<br>'+'Descrição: '+data.iloc[i]['desc']+'<br>'
    url = 'http://maps.google.com/maps?q=&layer=c&cbll='+str(data.iloc[i]['lat'])+','+str(data.iloc[i]['long'])
    popup += '<a target="_blank" href="'+url+'"> Vista da OAE </a>'
    fl.Marker([data.iloc[i]['lat'], data.iloc[i]['long']], popup=popup).add_to(marker_cluster)
```
